```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('dark_background')

import numpy as np
import jax
import jax.numpy as jnp
from ott.geometry.costs import CostFn

from utils import manifold, metrics
from sklearn.preprocessing import StandardScaler, MaxAbsScaler

from functools import partial
```

```python
def preprocess(data):
    return data

file_name = "double_pendulum_1000_500_v0.5"
data = np.load("data/" + file_name + ".npz")
data, params = data["data"], data["params"]

E = params[:,0]
E1 = params[:,1]
E2 = params[:,2]

data = preprocess(data)
```

```python
plt.figure()
for traj in data[:10]:
    plt.scatter(traj[:,0], traj[:,1], s=1)
plt.xlabel(r"angle $\theta$")
plt.ylabel(r"angular momentum $L$")
plt.axis('equal')

plt.figure()
for traj in data[:10]:
    plt.scatter(traj[:,2], traj[:,3], s=1)
plt.xlabel(r"angle $\theta$")
plt.ylabel(r"angular momentum $L$")
plt.axis('equal')

plt.figure()
plt.scatter(E+3, E1+E2, s=1);
```

```python
@jax.tree_util.register_pytree_node_class
class PeriodicEuclidean(CostFn):
    """Squared Periodic Euclidean distance CostFn."""
    def __init__(self, periods):
        super().__init__()
        self._periods = periods

    def pairwise(self, x, y):
        s = jnp.mod(jnp.abs(x - y), self._periods)
        s = jnp.minimum(s, self._periods - s)
        return jnp.sum(s ** 2)

    def tree_flatten(self):
        return (), (self._periods,)

    @classmethod
    def tree_unflatten(cls, aux_data, children):
        del children
        return cls(aux_data[0])

cost_fn = PeriodicEuclidean(jnp.array([2*jnp.pi, jnp.inf, 2*jnp.pi, jnp.inf]))
metric = partial(metrics.wasserstein_metric, cost_fn=cost_fn)
dist_mat, converged, steps = metrics.distance_matrix(data, metric=metric, batch_size=200, mesh_shape=(4,2))

print("Converged?", jnp.all(converged))
print(np.stack(np.where(1-converged), axis=1))

plt.figure()
plt.hist([steps.reshape(-1)], log=True)
plt.figure()
plt.imshow(dist_mat)
plt.colorbar();
```
